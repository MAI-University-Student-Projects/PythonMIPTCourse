Класс должен поддерживать основные математические операции (+, -, \*, /) за счет перегрузки соответствующих магических методов. Также он должен поддерживать возможность получить число в алгебраической и полярной форме. Допускается использование модуля `math`.

````python
import math
from functools import reduce
````

````python
class ComplexNumber:
    def __init__(self, x, y, polar = False):
        if polar:
            self._re = x * math.cos(y)
            self._im = x * math.sin(y)
        else:
            self._re = x
            self._im = y
    
    def __add__(self, z_oth):
        res = ComplexNumber(0, 0)
        res._re = self._re + z_oth._re
        res._im = self._im + z_oth._im
        return res
    
    def __sub__(self, z_oth):
        res = ComplexNumber(0, 0)
        res._re = self._re - z_oth._re
        res._im = self._im - z_oth._im
        return res
    
    def __mul__(self, z_oth):
        res = ComplexNumber(0, 0)
        res._re = self._re * z_oth._re - self._im * z_oth._im
        res._im = self._re * z_oth._im + self._im * z_oth._re
        return res
    
    def __truediv__(self, z_oth):
        res = ComplexNumber(0, 0)
        res._re = (self._re * z_oth._re + self._im * z_oth._im) / (z_oth._re ** 2 + z_oth._im ** 2)
        res._im = (self._im * z_oth._re -  self._re * z_oth._im) / (z_oth._re ** 2 + z_oth._im ** 2)
        return res
    
    def __eq__(self, z_oth):
        return (self._re == z_oth._re) and (self._im == z_oth._im)
    
    def get_complex(self, polar = False):
        res = (0.0, 0.0)
        if polar:
            res = (math.sqrt(self._re ** 2 + self._im ** 2), math.atan2(self._im, self._re))
        else:
            res = (self._re, self._im)
        return res

    def __str__(self):
        return f"{self._re:.2f} + {self._im:.2f}i"
    #get в алгебр и полярной форме
````

````python
z1 = ComplexNumber(3.4, 3.8)
z2 = ComplexNumber(1, 2.7)

z_s = ComplexNumber(4.4, 6.5)
z_dif = ComplexNumber(2.4, 1.1)
z_mul = ComplexNumber(-6.86, 12.98)
z_div = ComplexNumber(1.65, -0.65)

zp = ComplexNumber(2.1, 4.3, polar=True)
z_n = ComplexNumber(-0.84, 1.92)

print("Op result(+):", z1 + z2, "; true result:", z_s)
print("Op result(-):", z1 - z2, "; true result:", z_dif)
print("Op result(*):", z1 * z2, "; true result(.2f):", z_mul)
print("Op result/:", z1 / z2, "; true result(.2f):", z_div)
print("Polar to normal(re, im):", zp.get_complex(), "; true normal:", z_n)
print("Polar(module, rad): ", zp.get_complex(polar=True))

#тесты
````

Обеспечьте поддержку следующих операций: сложение, вычитание (с созданием нового вектора-результата), скалярное произведение, косинус угла, евклидова норма. Все операции, которые можно перегрузить с помощью магических методов, должны быть реализованы именно через них. Класс должен производить проверку консистентности аргументов для каждой операции и в случаях ошибок выбрасывать исключение `ValueError` с исчерпывающим объяснением ошибки.

````python
class Vector:
    def __init__(self, vector_values_list):
        self._crd = vector_values_list if len(vector_values_list) > 0 else (0, 0)
        self._length = len(self._crd)
        
    def __consistancy_check(self, oth):
        if not isinstance(oth, Vector):
            print(f"Non vector argument")
            raise ValueError
        if not self._length == oth._length:
            print(f"Different dimentions")
            raise ValueError
    
    def __add__(self, oth):
        self.__consistancy_check(oth)
        res = None
        try:
            res = [a + b for a, b in zip(self._crd, oth._crd)]
        except:
            print(f"op+ not available for some element type")
            raise ValueError
        return Vector(res)
    
    def __sub__(self, oth):
        self.__consistancy_check(oth)
        res = None
        try:
            res = [a - b for a, b in zip(self._crd, oth._crd)]
        except:
            print(f"op- not available for some element type")
            raise ValueError
        return Vector(res)
    
    def __str__(self):
        return f"{self._crd}"
        
    def inner_product(self, oth):
        self.__consistancy_check(oth)
        res = None
        try:
            res = sum((a * b for a, b in zip(self._crd, oth._crd)))
        except:
            print(f"op* not available for some element type")
            raise ValueError
        return res
    
    def norm(self):
        return math.sqrt(self.inner_product(self))
    
    def angle(self, oth):
        return math.acos(self.inner_product(oth) / self.norm() / oth.norm() )
````

````python
a = Vector([1, 2.4, 5, 79, 7])
b = Vector([2, 3.5, 1.2, 5, 8])

v_s = Vector([3, 5.9, 6.2, 84, 15])
v_d = Vector([1, 1.1, -3.8, -74, 1])

print("Op result(+):", a + b, "; true result:", v_s)
print("Op result(+):", b - a, "; true result:", v_d)
print("Norm a:", a.norm())
print("Angle a^b(rad):", a.angle(b))
print("<a, b>:", a.inner_product(b))
````

__Задание 4 (2 балл):__ Опишите декоратор, который принимает на вход функцию и при каждом её вызове печатает строку "This function was called N times", где N - число раз, которое это функция была вызвана на текущий момент (пока функция существует как объект, это число, очевидно, может только неубывать).

````python
#https://stackoverflow.com/questions/44968004/python-decorators-count-function-call
def calls_counter(func):
    def counter(*args, **kw):
        # function attribute (because functions are objects in python^ they have own attributes)
        counter.calls += 1
        return func(*args, **kw)
    counter.calls = 0 # initialize function attrivute
    return counter # return function (as object)

@calls_counter
def hello_world():
    print("Hello world")

print(hello_world.calls)    

hello_world()
hello_world()
hello_world()

print(hello_world.calls)
````

Обе процедуры должны быть реализованы с помощью шаблона "примесь" (Mixin), т.е. указанные функциональности должны быть описаны в специализированных классах.

В реализации математических операций запрещается пользоваться любыми функциями, требующими использования оператора `import`.

````python
class Matrix:
    def __init__(self, *args): # tuple of arguments
        if len(args) == 2 or len(args) == 3:
            self.__create_from_num(args)
        elif len(args) == 1:
            try:
                self.__create_from_list_of_lists(*args) # unpacking tuple of arguments
            except TypeError:
                self.__create_from_list(*args)
        else:
            raise ValueError
    
    def __create_from_list_of_lists(self, list_of_lists_from):
        n_tmp = len(list_of_lists_from[0])
        for l in list_of_lists_from:
            if len(l) != n_tmp:
                raise ValueError
        self._m = len(list_of_lists_from)
        self._n = n_tmp
        self._data = list_of_lists_from
        
    
    def __create_from_list(self, list_from):
        self._m = 1
        self._n = len(list_from)
        self._data = [list_from]
    
    def __create_from_num(self, tuple_from):
        self._m = tuple_from[0]
        self._n = tuple_from[1]
        val = tuple_from[2] if len(tuple_from) == 3 else 0
        self._data = [ [val if i == j else 0 for j in range(self._n)] for i in range(self._m) ]
        
    def __str__(self):
        return '\n'.join( '\t'.join(map(str, row)) for row in self._data )
    
    def __getitem__(self, idx):
        return self._data[idx] if idx < self._m else self._data[-1]
    
    def __setitem__(self, idx, value):
        if not isinstance(value, list):
            print("Cannot assign non-iterable to raw")
        self._data[idx] = value
    
    def __dimension_check(self, oth):
        if not isinstance(oth, Matrix):
            print(f"Non vector argument")
            raise ValueError
        if not ( self._m == oth._m and self._n == oth._n):
            print("Different dimensions")
            raise ValueError
    
    def __add__(self, oth):
        self.__dimension_check(oth)
        res = None
        try:
            res = [ [self[i][j] + oth[i][j] for j in range(self._n)] for i in range(self._m) ]
        except:
            print(f"op+ not available for some element type")
            raise ValueError
        return Matrix(res)
    
    def __sub__(self, oth):
        self.__dimension_check(oth)
        res = None
        try:
            res = [ [self[i][j] - oth[i][j] for j in range(self._n)] for i in range(self._m) ]
        except:
            print(f"op- not available for some element type")
            raise ValueError
        return Matrix(res)
    
    def __mul__(self, oth):
        self.__dimension_check(oth)
        res = None
        try:
            res = [ [self[i][j] * oth[i][j] for j in range(self._n)] for i in range(self._m) ]
        except:
            print(f"op* not available for some element type")
            raise ValueError
        return Matrix(res)
    
    def __truediv__(self, oth):
        self.__dimension_check(oth)
        res = None
        try:
            res = [ [self[i][j] / oth[i][j] for j in range(self._n)] for i in range(self._m) ]
        except:
            print(f"op/ not available for some element type")
            raise ValueError
        return Matrix(res)
    
    def dot(self, oth):
        if not self._n == oth._m or not isinstance(oth, Matrix):
            print(f"op dot available: dimensions error or not matrix")
            raise ValueError
        res_mtx = Matrix(self._m, oth._n)
        try:
            for i in range(self._m):
                for j in range(oth._n):
                    for k in range(self._n):
                        res_mtx[i][j] += self[i][k] * oth[k][j];
        except:
            print(f"op dot not available for some element type")
            raise ValueError
        return res_mtx

    def transpose(self):
        res_mtx = Matrix(self._n, self._m)
        for i in range(self._n):
            for j in range(self._m):
                res_mtx[i][j] = self[j][i]
        return res_mtx
    
    def tr(self):
        dim_lim = self._m if self._m < self._n else self._n
        return sum( (self[i][i] for i in range(dim_lim)) )
    
    def det(self):
        res = None
        try:
            u = self.lup()[1]
            res = reduce(lambda x, y: x * y, (u[i][i] for i in range(u._n)) )
        except ValueError:
            pass
        return res
    
    def lup(self):
        if self._m != self._n:
            print("non-quadratic matrix dimentions")
            raise ValueError
        
        l, u, p = Matrix(self._n, self._n), Matrix(self._data), Matrix(self._n, self._n, 1)
        for j in range(self._n):
            pivot_val = 0.0
            pivot_row = -1
            for i in range(j, self._n):
                tmp = abs(u[i][j])
                if (tmp - pivot_val) > 1e-8:
                    pivot_val = tmp
                    pivot_row = i
            if pivot_row == -1:
                print("singular matrix, cannot do LU(LUP)")
                raise ValueError
            l[j], l[pivot_row] = l[pivot_row], l[j]
            u[j], u[pivot_row] = u[pivot_row], u[j]
            p[j], p[pivot_row] = p[pivot_row], p[j]
            l[j][j] = 1
            for i in range (j+1, self._n):
                l[i][j] = u[i][j] / u[j][j]
                for k in range(j, self._n):
                    u[i][k] = u[i][k] - l[i][j] * u[j][k]
        return l, u, p
    
````
